```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler # MinMaxScaler can also be used
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import ConvLSTM2D, Dense, Flatten, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam

import optuna # Thư viện để tinh chỉnh siêu tham số

# --- 1. Load and Basic Preprocessing ---
try:
    df_full = pd.read_csv('/content/coin_Bitcoin.csv')
except FileNotFoundError:
    print("Error: coin_Bitcoin.csv not found. Please ensure the file is in the correct path.")
    exit()

df_full['Date'] = pd.to_datetime(df_full['Date'])
df_full.sort_values('Date', inplace=True)
df_full.set_index('Date', inplace=True)

# QUAN TRỌNG: Chỉ sử dụng 'Close' làm feature và target
feature_col = 'Close' # Feature duy nhất
target_col = 'Close'  # Chúng ta sẽ dự đoán giá Close

# Tạo DataFrame chỉ với cột 'Close'
df_processed = df_full[[feature_col]].copy()
df_processed.fillna(method='ffill', inplace=True) # Xử lý NaN đơn giản
df_processed.dropna(inplace=True) # Đảm bảo không còn NaN

if df_processed.empty:
    print("DataFrame is empty after preprocessing.")
    exit()

# --- 2. Feature Scaling ---
# Scale cột 'Close' (sẽ được dùng cho cả input và target)
scaler_close = StandardScaler() # Hoặc MinMaxScaler()
df_processed['Close_Scaled'] = scaler_close.fit_transform(df_processed[[feature_col]])

# --- 3. Function to Create Sequences for ConvLSTM ---
def create_sequences_convlstm_single_feature(data_scaled_feature_col, n_timesteps):
    """
    Tạo sequences cho ConvLSTM từ một feature đã scale (ví dụ: 'Close_Scaled').
    Input shape: (samples, timesteps, rows=1, cols=1, channels=1)
    """
    X_seq, y_seq = [], []
    # Dữ liệu feature (ví dụ: Close_Scaled)
    feature_values = data_scaled_feature_col.values

    for i in range(n_timesteps, len(feature_values)):
        # Mỗi sequence X là một chuỗi các giá trị feature của n_timesteps ngày trước đó
        sequence_data = feature_values[i-n_timesteps:i] # Shape: (n_timesteps,)
        X_seq.append(sequence_data)
        y_seq.append(feature_values[i]) # Target là giá trị feature của ngày hiện tại

    X_seq = np.array(X_seq)
    y_seq = np.array(y_seq)

    # Reshape X_seq cho ConvLSTM: (samples, timesteps_outer, rows, cols, channels)
    # Ở đây, mỗi timestep trong n_timesteps là một "timestep_outer".
    # "Hình ảnh" tại mỗi timestep_outer là một giá trị vô hướng (1x1).
    # rows = 1
    # cols = 1 (vì chỉ có 1 feature)
    # channels = 1
    # Shape cuối cùng: (samples, n_timesteps, 1, 1, 1)
    X_seq_reshaped = X_seq.reshape(X_seq.shape[0], n_timesteps, 1, 1, 1)

    return X_seq_reshaped, y_seq

# --- 4. Define the Objective Function for Optuna ---
split_ratio = 0.8
split_idx_global = int(len(df_processed) * split_ratio)

# Chia df_processed (chứa 'Close' và 'Close_Scaled')
df_train_global_processed = df_processed.iloc[:split_idx_global]
df_test_global_processed = df_processed.iloc[split_idx_global:]


def objective(trial):
    # --- Hyperparameters to tune ---
    n_timesteps = trial.suggest_int('n_timesteps', 10, 60, step=5) # Số ngày trong quá khứ
    filters_1 = trial.suggest_categorical('filters_1', [32, 64, 128])
    # kernel_size_col is now fixed to 1 as we only have 1 feature
    filters_2 = trial.suggest_categorical('filters_2', [16, 32, 64])
    dropout_rate = trial.suggest_float('dropout_rate', 0.1, 0.5, step=0.1)
    dense_units = trial.suggest_categorical('dense_units', [32, 64, 128])
    learning_rate = trial.suggest_float('learning_rate', 1e-4, 1e-2, log=True)
    batch_size = trial.suggest_categorical('batch_size', [16, 32, 64])

    # --- Create sequences based on n_timesteps ---
    # Sử dụng cột 'Close_Scaled' từ df_train_global_processed
    X_train, y_train = create_sequences_convlstm_single_feature(
        df_train_global_processed['Close_Scaled'], n_timesteps
    )

    val_split_index = int(len(X_train) * 0.85) # 15% cho validation
    X_train_fold, X_val_fold = X_train[:val_split_index], X_train[val_split_index:]
    y_train_fold, y_val_fold = y_train[:val_split_index], y_train[val_split_index:]

    if X_train_fold.shape[0] == 0 or X_val_fold.shape[0] == 0:
        print(f"Skipping trial due to empty train/val fold with n_timesteps={n_timesteps}")
        return float('inf')

    # --- Build the ConvLSTM model ---
    # Input shape: (n_timesteps, 1, 1, 1)
    model = Sequential()
    model.add(ConvLSTM2D(
        filters=filters_1,
        kernel_size=(1, 1), # (row_kernel, col_kernel) - col_kernel is 1
        activation='relu',
        input_shape=(n_timesteps, 1, 1, 1), # rows=1, cols=1 (single feature)
        padding='same',
        return_sequences=True
    ))
    model.add(BatchNormalization())
    model.add(Dropout(dropout_rate))

    model.add(ConvLSTM2D(
        filters=filters_2,
        kernel_size=(1, 1), # col_kernel is 1
        activation='relu',
        padding='same',
        return_sequences=False # Lớp ConvLSTM cuối cùng
    ))
    model.add(BatchNormalization())
    model.add(Dropout(dropout_rate))

    model.add(Flatten())
    model.add(Dense(dense_units, activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1))

    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='mse')

    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=0)

    history = model.fit(
        X_train_fold, y_train_fold,
        epochs=30,
        batch_size=batch_size,
        validation_data=(X_val_fold, y_val_fold),
        callbacks=[early_stopping],
        verbose=0
    )

    val_loss = min(history.history['val_loss'])
    trial.report(val_loss, step=0)
    if trial.should_prune():
        raise optuna.exceptions.TrialPruned()
    return val_loss

# --- 5. Run Optuna Study ---
study_name = "convlstm_bitcoin_tuning_close_only"
storage_name = f"sqlite:///{study_name}.db"

study = optuna.create_study(
    study_name=study_name,
    direction='minimize',
    storage=storage_name,
    load_if_exists=True,
    pruner=optuna.pruners.MedianPruner(n_startup_trials=5, n_warmup_steps=5, interval_steps=1)
)

n_trials = 30 # Giảm số trials cho ví dụ, tăng lên để có kết quả tốt hơn
study.optimize(objective, n_trials=n_trials, timeout=3600*1) # Giới hạn thời gian chạy 1 giờ

print("\n--- Optuna Study Results ---")
print(f"Number of finished trials: {len(study.trials)}")
if len(study.trials) > 0 and study.best_trial:
    best_trial = study.best_trial
    print("Best trial:")
    print(f"  Value (min val_loss): {best_trial.value}")
    print("  Params: ")
    for key, value in best_trial.params.items():
        print(f"    {key}: {value}")

    # --- 6. Train Final Model with Best Hyperparameters ---
    best_params = best_trial.params
    n_timesteps_best = best_params['n_timesteps']

    X_train_final, y_train_final = create_sequences_convlstm_single_feature(
        df_train_global_processed['Close_Scaled'], n_timesteps_best
    )
    X_test_final, y_test_final = create_sequences_convlstm_single_feature(
        df_test_global_processed['Close_Scaled'], n_timesteps_best
    )

    # Index cho y_test_final để vẽ đồ thị
    # y_test_original_index sẽ là index của df_test_global_processed sau khi bỏ đi n_timesteps_best phần tử đầu
    y_test_original_index = df_test_global_processed.index[n_timesteps_best:]


    print(f"X_train_final shape: {X_train_final.shape}, y_train_final shape: {y_train_final.shape}")
    print(f"X_test_final shape: {X_test_final.shape}, y_test_final shape: {y_test_final.shape}")

    if X_train_final.shape[0] == 0 or X_test_final.shape[0] == 0:
        print("Final training or test set is empty. Check n_timesteps_best and data splitting.")
        exit()

    model_final = Sequential()
    model_final.add(ConvLSTM2D(
        filters=best_params['filters_1'],
        kernel_size=(1, 1),
        activation='relu',
        input_shape=(n_timesteps_best, 1, 1, 1),
        padding='same', return_sequences=True
    ))
    model_final.add(BatchNormalization())
    model_final.add(Dropout(best_params['dropout_rate']))
    model_final.add(ConvLSTM2D(
        filters=best_params['filters_2'],
        kernel_size=(1, 1),
        activation='relu', padding='same', return_sequences=False
    ))
    model_final.add(BatchNormalization())
    model_final.add(Dropout(best_params['dropout_rate']))
    model_final.add(Flatten())
    model_final.add(Dense(best_params['dense_units'], activation='relu'))
    model_final.add(Dropout(best_params['dropout_rate']))
    model_final.add(Dense(1))

    optimizer_final = Adam(learning_rate=best_params['learning_rate'])
    model_final.compile(optimizer=optimizer_final, loss='mse')
    model_final.summary()

    early_stopping_final = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True, verbose=1)

    history_final = model_final.fit(
        X_train_final, y_train_final,
        epochs=50, # Huấn luyện lâu hơn với tham số tốt nhất
        batch_size=best_params['batch_size'],
        validation_split=0.15,
        callbacks=[early_stopping_final],
        verbose=1
    )

    # --- 7. Evaluate Final Model ---
    y_pred_train_scaled_final = model_final.predict(X_train_final)
    y_pred_test_scaled_final = model_final.predict(X_test_final)

    # Inverse transform
    y_pred_train_final = scaler_close.inverse_transform(y_pred_train_scaled_final)
    y_pred_test_final = scaler_close.inverse_transform(y_pred_test_scaled_final)

    y_train_orig_final = scaler_close.inverse_transform(y_train_final.reshape(-1, 1))
    # y_test_orig_final cần lấy từ df_processed gốc, không phải y_test_final (đã scale)
    # y_test_orig_final = df_test_global_processed[target_col].values[n_timesteps_best:]
    # Hoặc, an toàn hơn, inverse transform y_test_final
    y_test_orig_final = scaler_close.inverse_transform(y_test_final.reshape(-1,1))


    print("\n--- Final ConvLSTM Model Evaluation (Close Price Only) ---")
    if len(y_train_orig_final) > 0 and len(y_pred_train_final) > 0 :
        rmse_train = np.sqrt(mean_squared_error(y_train_orig_final, y_pred_train_final))
        mae_train = mean_absolute_error(y_train_orig_final, y_pred_train_final)
        r2_train = r2_score(y_train_orig_final, y_pred_train_final)
        mape_train = mean_absolute_percentage_error(y_train_orig_final, y_pred_train_final) * 100
        print(f"Final Training RMSE: ${rmse_train:.2f}, MAE: ${mae_train:.2f}, R2: {r2_train:.4f}, MAPE: {mape_train:.2f}%")
    else:
        print("Not enough data for training evaluation.")

    if len(y_test_orig_final) > 0 and len(y_pred_test_final) > 0:
        rmse_test = np.sqrt(mean_squared_error(y_test_orig_final, y_pred_test_final))
        mae_test = mean_absolute_error(y_test_orig_final, y_pred_test_final)
        r2_test = r2_score(y_test_orig_final, y_pred_test_final)
        mape_test = mean_absolute_percentage_error(y_test_orig_final, y_pred_test_final) * 100
        print(f"Final Test RMSE: ${rmse_test:.2f}, MAE: ${mae_test:.2f}, R2: {r2_test:.4f}, MAPE: {mape_test:.2f}%")
    else:
        print("Not enough data for test evaluation.")


    # --- 8. Visualize Predictions ---
    if len(y_test_original_index) == len(y_test_orig_final) and len(y_test_original_index) == len(y_pred_test_final):
        plt.figure(figsize=(15, 7))
        plt.plot(y_test_original_index, y_test_orig_final, label='Test Data (Actual Close)', color='green', alpha=0.7)
        plt.plot(y_test_original_index, y_pred_test_final, label='Test Data (Predicted Close by Tuned ConvLSTM)', color='red', linestyle='--')
        plt.title('Bitcoin Close Price Prediction using Tuned ConvLSTM (Close Only)')
        plt.xlabel('Date')
        plt.ylabel('Price (USD)')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    else:
        print("Length mismatch for plotting test predictions. Skipping plot.")
        print(f"Len y_test_original_index: {len(y_test_original_index)}")
        print(f"Len y_test_orig_final: {len(y_test_orig_final)}")
        print(f"Len y_pred_test_final: {len(y_pred_test_final)}")


    plt.figure(figsize=(10, 5))
    plt.plot(history_final.history['loss'], label='Training Loss')
    plt.plot(history_final.history['val_loss'], label='Validation Loss')
    plt.title('Final Model Loss During Training (Close Only)')
    plt.xlabel('Epoch')
    plt.ylabel('Loss (MSE)')
    plt.legend()
    plt.grid(True)
    plt.show()

    # --- 9. Make a sample forecast for the next day ---
    last_n_days_scaled_close = df_processed['Close_Scaled'].iloc[-n_timesteps_best:].values
    next_day_input_reshaped = last_n_days_scaled_close.reshape((1, n_timesteps_best, 1, 1, 1))
    next_day_pred_scaled = model_final.predict(next_day_input_reshaped)
    next_day_prediction = scaler_close.inverse_transform(next_day_pred_scaled)

    last_actual_date_in_df_full = df_full.index[-1]
    next_prediction_date = last_actual_date_in_df_full + pd.Timedelta(days=1)

    print(f"\n--- Tuned ConvLSTM Next Day Forecast (Close Only) ---")
    print(f"Last actual Close on {df_full[target_col].index[-1].date()}: ${df_full[target_col].iloc[-1]:.2f}")
    print(f"Past {n_timesteps_best} Close prices were used as input.")
    print(f"Predicted Close by Tuned ConvLSTM for {next_prediction_date.date()}: ${next_day_prediction[0, 0]:.2f}")

    # model_final.save(f"tuned_convlstm_bitcoin_model_close_only.h5")
    # print("Final model saved.")
else:
    print("Optuna study did not complete any trials or find a best trial.")

# Để xem lại các trials của Optuna:
# optuna-dashboard sqlite:///convlstm_bitcoin_tuning_close_only.db
```
